### 1) Dataset Selection:  
 

```python
import pandas as pd
```

```python
df = pd.read_csv('WorldCupMatches.csv')
```

```python
df.head()
```

### 2) Data Wrangling and Cleaning: 

```python
df.info()
```

```python
df.isnull().sum()
```

```python
numeric_columns = df.select_dtypes(include='number').columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())
```

```python
categorical_columns = df.select_dtypes(include='object').columns
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])
```

```python
df.isnull().sum()
```

### 3) Descriptive Statistics:  

```python
print("Descriptive Stats: ")
df.describe()
```

```python
print("Mode Stats: ")
df.mode()
```

### 4) Data Visualization:

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Stage', y='Attendance', data=df)
plt.title('Box Plot of Attendance by Stage')
plt.xlabel('Stage')
plt.ylabel('Attendance')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Home Team Goals'], bins=20, kde=True)
plt.title('Histogram of Home Team Goals')
plt.xlabel('Home Team Goals')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Home Team Goals', y='Away Team Goals', data=df)
plt.title('Scatter Plot of Home Team Goals vs Away Team Goals')
plt.xlabel('Home Team Goals')
plt.ylabel('Away Team Goals')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Stage', y='Home Team Goals', data=df)
plt.title('Box Plot of Home Team Goals by Stage')
plt.xlabel('Stage')
plt.ylabel('Home Team Goals')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Stage', y='Attendance', data=df)
plt.title('Box Plot of Attendance by Stage')
plt.xlabel('Stage')
plt.ylabel('Attendance')
plt.xticks(rotation=45)
plt.show()
```
